```python
w1 = np.loadtxt('mnist_edge10_SQA/metropolis_20/w1_0.txt')
w2 = np.loadtxt('mnist_edge10_SQA/metropolis_20/w2_0.txt')
w3 = np.loadtxt('mnist_edge10_SQA/metropolis_20/w3_0.txt')
w4 = np.loadtxt('mnist_edge10_SQA/metropolis_20/w4_0.txt')
m1_ = np.loadtxt('mnist_edge10_SQA/metropolis_20/m1_reduce10.0.txt')
m2_ = np.loadtxt('mnist_edge10_SQA/metropolis_20/m2_reduce10.0.txt')
m3_ = np.loadtxt('mnist_edge10_SQA/metropolis_20/m3_reduce10.0.txt')
m4_ = np.loadtxt('mnist_edge10_SQA/metropolis_20/m4_reduce10.0.txt')

w1 = w1.flatten()
w2 = w2.flatten()
w3 = w3.flatten()
w4 = w4.flatten()
m1_ = m1_.flatten()
m2_ = m2_.flatten()
m3_ = m3_.flatten()
m4_ = m4_.flatten()

w = np.hstack([w1,w2,w3,w4])
m = np.hstack([m1_,m2_,m3_,m4_])

w_reduce = []

for i in range(len(w)):
    if m[i] == 1:
        w_reduce.append(w[i])
        
plt.figure(figsize=(8,5))
plt.hist(w,bins=100,range=[-0.3,0.2],color='lightsteelblue',label='Inital')
plt.hist(w_reduce,bins=100,range=[-0.3,0.2],color='cornflowerblue',label='Remained')

plt.xlabel('weights',fontsize=18)
plt.ylabel('Count',fontsize=18)
# plt.title('10% Pruning')
plt.legend(fontsize=18)
plt.savefig('10_pruning_hist_large_network.pdf')
```
